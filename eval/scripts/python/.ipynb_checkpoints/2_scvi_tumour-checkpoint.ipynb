```python
# integration with scVI
adata = adata.copy()
scvi.model.SCVI.setup_anndata(
    adata = adata, 
    batch_key = "sample"
)
# create model
scvi_model = scvi.model.SCVI(
    adata = adata,
    n_hidden = 128,
    n_latent = 50,
    dispersion = "gene-batch"
)
# train model 
# model.train()
# scvi.settings.dl_num_workers = 11
scvi_model.train(
    max_epochs = 1000, 
    early_stopping = True,
    accelerator = "gpu"
)
```

```python
scvi_model.history["elbo_validation"].plot()
```

```python
SCVI_LATENT_KEY = "X_scVI"
adata.obsm[SCVI_LATENT_KEY] = scvi_model.get_latent_representation()
```

```python
pdn_index = (adata.obs["type"] == "PDN")
pdo_index = (adata.obs["type"] == "PDO")
t_index = (adata.obs["type"] == "T")

res_pdn_t = scvi_model.differential_expression(
    adata = adata, 
    idx1 = pdn_index,
    idx2 = t_index,
    weights = "uniform",
    # batch_correction = True
)
res_pdo_t = scvi_model.differential_expression(
    adata = adata, 
    idx1 = pdo_index,
    idx2 = t_index,
    weights = "uniform",
    # batch_correction = True
)

res_pdn_t.to_csv(path_or_buf = "/vast/scratch/users/moore.z/pdo_compar/data/processed/broad_pdn_t.csv")
res_pdo_t.to_csv(path_or_buf = "/vast/scratch/users/moore.z/pdo_compar/data/processed/broad_pdo_t.csv")
```

```python
per_anno_pdo_tis = pd.DataFrame()
for i in list(adata.obs["individual_anno"].unique()):
    
    idx_1 = (adata.obs["type"] == "PDO") & (adata.obs["individual_anno"] == i)
    idx_2 = (adata.obs["type"] == "T") & (adata.obs["individual_anno"] == i)
    
    if not any(idx_1) or not any(idx_2):
        continue
    
    res_pdo_tis = scvi_model.differential_expression(
        adata = adata, 
        idx1 = idx_1,
        idx2 = idx_2,
        filter_outlier_cells = False,
        batch_correction = False
    )
    
    res_pdo_tis.insert(0, "individual_anno", i)
    
    per_anno_pdo_tis = pd.concat([per_anno_pdo_tis, res_pdo_tis])
    
per_anno_pdn_tis = pd.DataFrame()
for i in list(adata.obs["individual_anno"].unique()):
    
    idx_1 = (adata.obs["type"] == "PDN") & (adata.obs["individual_anno"] == i)
    idx_2 = (adata.obs["type"] == "T") & (adata.obs["individual_anno"] == i)
    
    if not any(idx_1) or not any(idx_2):
        continue
    
    res_pdn_tis = scvi_model.differential_expression(
        adata = adata, 
        idx1 = idx_1,
        idx2 = idx_2,
        filter_outlier_cells = False,
        batch_correction = False
    )
    
    res_pdn_tis.insert(0, "individual_anno", i)
    
    per_anno_pdn_tis = pd.concat([per_anno_pdn_tis, res_pdn_tis])
```

```python
per_anno_pdn_pdo = pd.DataFrame()
for i in list(adata.obs["individual_anno"].unique()):
    
    idx_1 = (adata.obs["type"] == "PDN") & (adata.obs["individual_anno"] == i)
    idx_2 = (adata.obs["type"] == "PDO") & (adata.obs["individual_anno"] == i)
    
    if not any(idx_1) or not any(idx_2):
        continue
    
    res_pdn_pdo = scvi_model.differential_expression(
        adata = adata, 
        idx1 = idx_1,
        idx2 = idx_2,
        filter_outlier_cells = False,
        batch_correction = False
    )
    
    res_pdn_pdo.insert(0, "individual_anno", i)
    
    per_anno_pdn_pdo = pd.concat([per_anno_pdn_pdo, res_pdn_pdo])
```

```python
scvi.model.SCVI.setup_anndata(
    reference_adata, 
    batch_key = "Sample", 
    layer = "counts"
)

scvi_reference_model = scvi.model.SCVI(
    adata = reference_adata,
    n_hidden = 128,
    n_latent = 50,
    dispersion = "gene-batch"
)
# train model 
# model.train()
# scvi.settings.dl_num_workers = 11
scvi_reference_model.train(
    max_epochs = 1000, 
    early_stopping = True
)
```

```python
scvi_reference_model.history["elbo_validation"].plot()
```

```python
SCVI_LATENT_KEY = "X_scVI"
reference_adata.obsm[SCVI_LATENT_KEY] = scvi_reference_model.get_latent_representation()
```

```python
sc.pp.neighbors(
    adata = reference_adata, 
    use_rep = SCVI_LATENT_KEY, 
    n_neighbors = 15,
    n_pcs = 50
)
sc.tl.leiden(
    reference_adata,
    flavor = "igraph",
    resolution = 1,
    n_iterations = 10
)
```
